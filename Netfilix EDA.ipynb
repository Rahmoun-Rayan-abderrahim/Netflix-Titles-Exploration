```python
import kagglehub
import shutil
import os

# Download original dataset to cache
path = kagglehub.dataset_download("shivamb/netflix-shows")

# Path of your notebook directory (current directory)
target_dir = os.getcwd()

# Copy ALL dataset files to the notebook folder
for file in os.listdir(path):
    shutil.copy(os.path.join(path, file), target_dir)

print("Dataset copied to:", target_dir)
```
